```python
a = [1,2,3]
b = [4,5,6]
c=[]

for i in a :
    for j in b:
        if i*j>=10:
            c.append((i, j, i*j))    
print(c)


 
c = [(i, j, i*j) for i in a for j in b if i*j >= 10]
print(c)
```

```python
sentences = ["I love Python", "Python is fun", "I am learning"]
word=[]

for i in sentences:
    if "Python" in i :
        for j in i.split():
            word.append(j)
print(word)

sentences = ["I love Python", "Python is fun", "I am learning"]
 
a = [j for i in sentences if "Python" in i for j in i.split()]
print(a)
```
