```python
# 1. words = ["I", "study", "python", "language", "!"] 데이터에 대한
#    아래 코드 작성하세요.
words = ["I", "study", "python", "language", "!"]
print('#----------------------------------')
print('# 1번')
print('#----------------------------------')

# (1) words의 모든 요소를 for반복문으로 출력
print('[1]---')
for i in words:
    print(i)

# (2) words에서 홀수번째 요소만 for반복문으로 출력
print('\n[2]---')
for i in range(0, len(words), 2):   #0번째 인덱스에서 words의 길이까지 2씩 증가하면서 반복
    print(words[i])

# (3) words에서 길이가 3이상인 요소만 for반복문으로 출력
print('\n[3]---')
for i in words:
    if len(i) >= 3:
        print(i)

# (4) words를 오름차순 정렬하여 모든 요소 출력
print('\n[4]---')
for i in sorted(words): #sorted() 함수를 사용하여 words 리스트를 오름차순으로 정렬한 후, for 반복문을 통해 각 요소를 출력
    print(i)
```

```python
# ## ----------------------------------------------------------------
# 2.  아래 데이터에 대한 구현 코드 작성하세요.
#     files = ['intra.h', 'intra.c', 'define.h', 'run.py', 'ex01.py', 'intro.hwp']
#     (1) 확장자 없이 파일명만 출력
#     (2) 확장자가 ‘h’이거나 ‘c’인 것만 출력
#     (3) 파일명에 ‘e’가 2개 이상 있고 ‘f’가 있는 파일명만 출력
print('\n\n#----------------------------------')
print('# 2번')
print('#----------------------------------')

files = ['intra.h', 'intra.c', 'define.h', 'run.py', 'ex01.py', 'intro.hwp']

# (1) 확장자 없이 파일명만 출력
print('[1]---')
for j in files:
    print(j.split('.')[0]) #split() 함수를 사용하여 파일명을 '.'을 기준으로 분리한 후, 첫 번째 요소(인덱스 0)를 출력

# (2) 확장자가 ‘h’이거나 ‘c’인 것만 출력
print('\n[2]---')
for j in files:
    확장자= j.split('.')[1]
    if 확장자== 'h' or 확장자== 'c':
        print(확장자)

# (3) 파일명에 ‘e’가 2개 이상 있고 ‘f’가 있는 파일명만 출력
print('\n[3]---')
for j in files:
    name = j.split('.')[0] #파일명을 '.'을 기준으로 분리한 후, 첫 번째 요소(인덱스 0)를 name 변수에 저장
    if name.count('e') >= 2 and 'f' in name: #count() 함수를 사용하여 파일명에서 'e'의 개수를 세고, 'f'가 파일명에 포함되어 있는지 확인
        print(j)
```

```python

# ## ----------------------------------------------------------------
# 3. for 반복문을 사용하여 2단부터 9단까지 모두 출력하세요.
print('\n\n#----------------------------------')
print('# 3번')
print('#----------------------------------')

for k in range(2, 10):
    print(f'\n    {k}')
    for n in range(1, 10):
        print(f'{k} * {n} = {k*n:2}') #f-string을 사용하여 구구단의 결과를 출력하는데, {k*n:2}는 k와 n의 곱을 2자리로 맞춰서 출력하도록 지정하는 형식입니다. 이를 통해 결과가 정렬되어 보이도록 합니다.

```

```python
# ## ----------------------------------------------------------------
# 4. 2개의 정수와 덧셈, 뺄셈, 곱셈, 나눗셈 연산자 1개를 입력 받으세요.
#    연산 결과를 출력합니다. 단, 입력 함수는 1개 사용합니다.
print('\n\n#----------------------------------')
print('# 4번')
print('#----------------------------------')

data = input('입력: ').split() #함수는 한 번만 써야해서 이렇게 쓰기
num1 = int(data[0])
op= data[1]
num2 = int(data[2]) 

if op== '+':
    print(num1+ num2) #
elif op== '-':
    print(num1- num2)
elif op== '*':
    print(num1* num2)
elif op== '/':
    print(num1/ num2)
else:
    print('잘못된 연산자입니다.')

```

```python
# ## ----------------------------------------------------------------
# 5. “가나*마바사*자***파하”에서 첫 번째 *의 인덱스를 출력하세요.
#    세 번째 ‘*’의 인덱스를 출력하세요.
print('\n\n#----------------------------------')
print('# 5번')
print('#----------------------------------')

s = "가나*마바사*자***파하"
print('첫번째 *위치:', s.find('*'))

count= 0
for i in range(len(s)):
    if s[i] == '*': #'*' 문자를 찾을 때마다 count 변수를 1씩 증가시킵니다.
        count+= 1
        if count== 3:
            print('세번째 *위치:', i) #count 변수를 사용하여 '*'의 개수를 세고, 세 번째 '*'이 나올 때 해당 인덱스를 출력하도록 합니다.
            break
```

```python
# ## ----------------------------------------------------------------
# 7. “1,234,567,890” 문자열에서 숫자들의 합계를 계산 후 출력하세요.
print('\n\n#----------------------------------')
print('# 7번')
print('#-------------------------------')

string= "1,234,567,890"
total= 0
for i in string:
    if i.isdigit(): #isdigit() 메서드를 사용하여 i가 숫자인지 확인합니다. 만약 i가 숫자라면, int() 함수를 사용하여 i를 정수로 변환한 후 total 변수에 더합니다.
        total+= int(i)
print(total)
```

```python
## ----------------------------------------------------------------
# 8. “산토끼 토끼야 어디를 가느냐 폴짝 폴짝 뛰면서 어디를 가느냐” 문장을
#     아래와 같이 출력하세요.
print('\n\n#----------------------------------')
print('# 8번')
print('#-------------------------------')

txt = "산토끼 토끼야 어디를 가느냐 폴짝 폴짝 뛰면서 어디를 가느냐"
words= txt.split()

for w in words:
    print(w[::-1], end=' ') #문자열을 뒤집어서 출력하는 방법입니다. w[::-1]은 문자열 w를 뒤집어서 반환합니다. end=' '는 각 단어 사이에 공백을 추가하여 출력하도록 지정하는 것입니다.
print()
```

```python
## ----------------------------------------------------------------
# 9. [9, 3, 1, 8, 7, 2, 1, 4, 2, 3, 5, 7] 에서 중복된 데이터 제외한
#     데이터들 평균 출력하세요.
print('\n\n#----------------------------------')
print('# 9번')
print('#-------------------------------')

nums= [9, 3, 1, 8, 7, 2, 1, 4, 2, 3, 5, 7]
nums1 = list(set(nums)) #set() 함수를 사용하여 nums 리스트에서 중복된 요소를 제거한 후, list() 함수를 사용하여 다시 리스트로 변환하여 nums1 변수에 저장합니다. 이렇게 하면 nums1에는 중복된 데이터가 제외된 데이터들만 포함됩니다.
avg = sum(nums1) / len(nums1)
print(avg)
```
