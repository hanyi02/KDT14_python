```python
a= 10, 20, 30

print(type(a)) #a는 튜플로 저장됩니다. 튜플은 변경할 수 없는 자료형입니다. 따라서 a의 타입은 <class 'tuple'>이 됩니다.
```

```python
a= range(-10, 10, 3)
b= list(range(-10, 10, 3))
#c= tuple(-10, 10, 3)
d= tuple(range(-10, 10, 3))
e= -10, 10, 3

print(a)
print(b)
#print(c)
print(d)
print(e)
```

```python
a=(11, 22, 33, 44, 55, 66, 77, 88, 99)
print(a[0:9])
print(a[0:9:3])
print(a[::4])
print(a[0:6:2])
print(a[0:len(a):4])
```

```python
x= input("숫자 또는 문자열을 입력하세요:").split()
del x[-5:] #del 키워드를 사용하여 리스트 x의 슬라이스 x[-5:]를 삭제
print(tuple(x)) #삭제된 리스트 x를 튜플로 변환하여 출력4
```

```python
s1= input("첫 번째 문자열 입력: ")
s2= input("두 번째 문자열 입력: ")
result= s1[1::2] + s2[0::2]
print(result)
```

```python
print('{0:>12.2f}'.format(1675.3) )
print('{0:>12}'.format(1675.3)  )
print('{0:>12.4f}'.format(1675.3)  )
print('{:>12.4f}'.format(1675.3)  )
print('{:0>12.4f}'.format(1675.3) )
```
