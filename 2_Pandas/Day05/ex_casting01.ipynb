-Dataframe/Series의 원소/요소의 실제 데이터에 맞게 타입 변환
 * 현재 타입 확인 속성 : dtype/dtypes
 * 현재 타입 변경 메서드 : 변수명.astype(변환_타입명) 반드시 저장 필수!!

```python
## 모듈 로딩
import pandas as pd             # 데이터 분석용 패키지
import numpy as np              # 배열을 파이썬 언어로 구현한 패키지
                                # pandas에서 데이터 저장을 numpy의 배열로 함!
```

```python
## 데이터 준비 및 DF 인스턴스 생성
df = pd.DataFrame({
"이름": ["김민수", "이서연", "박지훈"],
"나이": ["20", "21", "19"],
"점수": ["85.5", "90.0", "78.5"],
"가입일": ["2024-01-10", "2024-02-15", "2024-03-01"]
})
```

```python
## DF 인스턴스 확인
df
df.index

## 실제 데이터 일부
df.head(2)

## > 실제 데이터와 타입이 맞지 않는 컬럼 확인
## 나이 >>> int 점수 >>> float, 가입일 >> datetime64[ns]
## >> 변환 자료형 : 타입명 입력 시 가장 큰 값의 타입 (int > int64, float > float64)
df['나이'] = df['나이'].astype(int)    # object 에서 int64로 변경된다, inplace매개변수 x, 그래서 반드시 저장 
df['점수'] = df['점수'].astype(float)
df['가입일'] = df['가입일'].astype('datetime64[ns]')    # 날짜시간타입명은 반드시 문자열!!
```

```python
## >> 변환 자료형 : 타입명 입력 시 가장 큰 값의 타입 (int > int64, float > float64)
## >> 결측치 처리 가능한 타입 : 첫글자 대문자
df['나이'] = df['나이'].astype('Int8')    # object 에서 int64로 변경된다, inplace매개변수 x, 그래서 반드시 저장
df['점수'] = df['점수'].astype('Float32')
df['가입일'] = df['가입일'].astype('datetime64[ns]')    # 날짜시간타입명은 반드시 문자열!! 권장은 X


df.dtypes
```

```python
print("변환 전=====\n", df.dtypes)
```

```python
## 가입일 컬럼 object >>> datetime64[ns] 형변환
pd.to_datetime(df.가입일)
```
